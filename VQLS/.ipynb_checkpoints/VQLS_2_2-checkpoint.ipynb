```python
import qiskit
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit import Aer, execute
import math
import random
import numpy as np
from scipy.optimize import minimize
```

```python
import numpy as np
I = np.asmatrix('1 0;0 1')
X = np.asmatrix('0 1;1 0')
Y = np.asmatrix('0 0-1j;0+1j 0')
Z = np.asmatrix('1 0;0 -1')
pauli_i = [I, 0]
pauli_x = [X, 1]
pauli_y = [Y, 2]
pauli_z = [Z, 3]

def get_basis_set(num_of_qubit = 1):
    basis_set = [pauli_i, pauli_x, pauli_y, pauli_z]
    int_set = [pauli_i, pauli_x, pauli_y, pauli_z]
    temp_set = []
    for i in range (num_of_qubit-1):
        for j in range (len(basis_set)):
            for k in range (len(int_set)):
                a = basis_set[j]
                b = int_set[k]
                #temp_set.append()
                temp_set.append([np.kron(a[0],b[0]),[a[1],b[1]]])
        int_set = temp_set
        #print(int_set)
        temp_set = []
    return int_set

def get_decomposition(A, num_of_q):
    basis = get_basis_set(num_of_q)
    dec_list = []
    for i in range(len(basis)):
        b = basis[i]
        coeff = np.trace(A*b[0])/(2**num_of_q)
        dec_list.append([coeff, b[1]])
    return dec_list
def get_arbitary_decomposition(A):
    size = A.shape
    size = size[0]
    power = math.log2(size)
    p = int(power)
    if (p != power):
        p += 1
    ret = np.zeros([2**p,2**p], dtype = float)
    ret[:size, :size] = A
    for i in range(size, 2**p):
        ret[i,i] = 1
    return get_decomposition(ret, p)
```

```python
backend = Aer.get_backend('qasm_simulator') 
T=8192
shots = T
max_credits = 3
def cal_cost(angle):
    den = 0 
    numer = 0
    qc = QuantumCircuit(2, 1)
    qc.h(0)
    qc.ry(angle[0], 1)
    qc.z(1)
    qc.ry(angle[0], 1)
    qc.cz(0, 1)
    qc.cz(0, 1)
    qc.h(0)
    qc.measure(0, 0)
    job = execute(qc, backend=backend, shots=shots,max_credits=max_credits)    
    result = job.result()
    counts = result.get_counts(qc)
    m_sum = 0
    if "0" in list (counts):
        m_sum=m_sum+counts["0"]/T
    if "1" in list (counts):
        m_sum=m_sum-counts["1"]/T
    den += 2*2*m_sum
    qc = QuantumCircuit(2, 1)
    qc.h(0)
    qc.ry(angle[0], 1)
    qc.z(1)
    qc.ry(angle[0], 1)
    qc.cz(0, 1)
    qc.cx(0, 1)
    qc.h(0)
    qc.measure(0, 0)
    job = execute(qc, backend=backend, shots=shots,max_credits=max_credits)    
    result = job.result()
    counts = result.get_counts(qc)
    m_sum = 0
    if "0" in list (counts):
        m_sum=m_sum+counts["0"]/T
    if "1" in list (counts):
        m_sum=m_sum-counts["1"]/T
    den += 2*0.5*m_sum
    qc = QuantumCircuit(2, 1)
    qc.h(0)
    qc.ry(angle[0], 1)
    qc.z(1)
    qc.ry(angle[0], 1)
    qc.cx(0, 1)
    qc.cz(0, 1)
    qc.h(0)
    qc.measure(0, 0)
    job = execute(qc, backend=backend, shots=shots,max_credits=max_credits)    
    result = job.result()
    counts = result.get_counts(qc)
    m_sum = 0
    if "0" in list (counts):
        m_sum=m_sum+counts["0"]/T
    if "1" in list (counts):
        m_sum=m_sum-counts["1"]/T
    den += 2*0.5*m_sum
    qc = QuantumCircuit(2, 1)
    qc.h(0)
    qc.ry(angle[0], 1)
    qc.z(1)
    qc.ry(angle[0], 1)
    qc.cx(0, 1)
    qc.cx(0, 1)
    qc.h(0)
    qc.measure(0, 0)
    job = execute(qc, backend=backend, shots=shots,max_credits=max_credits)    
    result = job.result()
    counts = result.get_counts(qc)
    m_sum = 0
    if "0" in list (counts):
        m_sum=m_sum+counts["0"]/T
    if "1" in list (counts):
        m_sum=m_sum-counts["1"]/T
    den += 0.5*0.5*m_sum
    
    qc = QuantumCircuit(3, 1)
    qc.h(0)
    qc.ry(angle[0], 1)
    qc.z(1)
    qc.ry(angle[0], 1)
    qc.cz(0, 1)
    qc.h(2)
    qc.cz(0, 2)
    qc.cx(1, 2)
    qc.h(0)
    qc.measure(0, 0)
    job = execute(qc, backend=backend, shots=shots,max_credits=max_credits)    
    result = job.result()
    counts = result.get_counts(qc)
    m_sum = 0
    if "0" in list (counts):
        m_sum=m_sum+counts["0"]/T
    if "1" in list (counts):
        m_sum=m_sum-counts["1"]/T
    numer += 2*2*m_sum
    qc = QuantumCircuit(3, 1)
    qc.h(0)
    qc.ry(angle[0], 1)
    qc.z(1)
    qc.ry(angle[0], 1)
    qc.cz(0, 1)
    qc.h(2)
    qc.cx(0, 2)
    qc.cx(1, 2)
    qc.h(0)
    qc.measure(0, 0)
    job = execute(qc, backend=backend, shots=shots,max_credits=max_credits)    
    result = job.result()
    counts = result.get_counts(qc)
    m_sum = 0
    if "0" in list (counts):
        m_sum=m_sum+counts["0"]/T
    if "1" in list (counts):
        m_sum=m_sum-counts["1"]/T
    numer += 2*0.5*m_sum
    qc = QuantumCircuit(3, 1)
    qc.h(0)
    qc.ry(angle[0], 1)
    qc.z(1)
    qc.ry(angle[0], 1)
    qc.cx(0, 1)
    qc.h(2)
    qc.cz(0, 2)
    qc.cx(1, 2)
    qc.h(0)
    qc.measure(0, 0)
    job = execute(qc, backend=backend, shots=shots,max_credits=max_credits)    
    result = job.result()
    counts = result.get_counts(qc)
    m_sum = 0
    if "0" in list (counts):
        m_sum=m_sum+counts["0"]/T
    if "1" in list (counts):
        m_sum=m_sum-counts["1"]/T
    numer += 2*0.5*m_sum
    qc = QuantumCircuit(3, 1)
    qc.h(0)
    qc.ry(angle[0], 1)
    qc.z(1)
    qc.ry(angle[0], 1)
    qc.cx(0, 1)
    qc.h(2)
    qc.cx(0, 2)
    qc.cx(1, 2)
    qc.h(0)
    qc.measure(0, 0)
    job = execute(qc, backend=backend, shots=shots,max_credits=max_credits)    
    result = job.result()
    counts = result.get_counts(qc)
    m_sum = 0
    if "0" in list (counts):
        m_sum=m_sum+counts["0"]/T
    if "1" in list (counts):
        m_sum=m_sum-counts["1"]/T
    numer += 2*2*m_sum
    cost = 1 - abs(numer/den)
    return abs(cost)
```
